## 1. What is Data Analysis?

**Data Analysis** is the process of collecting, cleaning, transforming, and interpreting data to extract meaningful insights and support decision-making.

### Key Steps in Data Analysis:

- **Data Collection** – Gathering raw data from various sources  
- **Data Cleaning** – Removing errors, duplicates, missing values  
- **Data Transformation** – Structuring data into usable format  
- **Data Visualization** – Using charts/graphs for understanding  
- **Conclusion** – Drawing insights for decision-making  

### Example:

A company analyzes customer purchase data to identify which products sell the most.

---

## 4. What are Tensors?

### Definition:
A **Tensor** is a multi-dimensional array used to represent data in machine learning and deep learning.

### Types of Tensors:

| Type   | Example           | Dimension |
|--------|------------------|----------|
| Scalar | 5                | 0D       |
| Vector | [1, 2, 3]        | 1D       |
| Matrix | [[1,2],[3,4]]    | 2D       |
| Tensor | 3D+ array        | ND       |

---

### Tensor Explanation with NumPy

```python
import numpy as np
```

```python
# 1. Scalar (0D Tensor)
a = np.array(10)
print(a)
print(a.shape)
```

```python
# 2. Vector (1D Tensor)
b = np.array([1,2,3])
print(b)
print(b.shape) 
```

```python
# 3. Matrix (2D Tensor)
c = np.array([[1,2],[3,4]])
print(c)
print(c.shape) 
```

```python
# 4. 3D Tensor
d = np.array([
    [[1,2],[3,4]],
    [[5,6],[7,8]]
])
print(d)
print(d.shape)
```

```python
print(d.ndim)   # Number of dimensions
print(d.shape)  # Size in each dimension
print(d.size)   # Total elements
```

### 4. Fetch Data from an API

```python
import requests 

url = "https://randomuser.me/api/?results=5"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    user = data['results']
    df = pd.json_normalize(user)
    print(df.head())
else : 
    print("Error fetching data")
    
```

# Part D: Exploratory Data Analysis (EDA)
> EDA helps in understanding patterns, trends, and distributions in the dataset.


## 8. Perform Univariate Analysis

```python
df = pd.read_csv("Data.csv")
df
```

```python
# Distribution Plot for Age
import matplotlib.pyplot as plt
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### What I Notices : 
- Range: 20 to 60 years  
- Highest peak at 59–60 (frequency 11) → majority are older customers  
- Second peak at 20–22 (frequency 8) → young customers also active  
- Age group 27–37 has lowest frequency → least represented  
- Distribution is right-skewed (more older customers)  
- Two dominant segments:
  - Young (20–22)
  - Older (55–60)

```python
# Distribution Plot for Income
plt.hist(df['Income'], bins=20)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

### What I Notices :
- Range: ₹22,000 to ₹93,000+  
- Peak at ₹40K–₹43K (frequency 11) → most common income group  
- Second peak at ₹22K–₹25K (frequency 8) → low-income group  
- Another rise at ₹70K–₹72K (frequency 8) → high-income group  
- Lowest at ₹27K–₹29K and ₹65K–₹67K → gaps in mid ranges  
- Distribution is **multimodal**  
- Three segments:
  - Low income
  - Middle income
  - High income  

```python
# Distribution Plot for Purchases
plt.hist(df['Purchases'], bins=20)
plt.title("Purchases Distribution")
plt.xlabel("Purchases")
plt.ylabel("Frequency")
plt.show()
```

### What I Notices :
- Range: 1 to 50 purchases  
- Peak at 1–3 purchases (frequency 8) → occasional buyers  
- Peak at 22–24 purchases (frequency 9) → regular buyers  
- High values at 40–44 purchases → loyal customers  
- Lowest at 11–13 purchases  
- Distribution is **bimodal**
- Two customer types:
  - Occasional buyers (1–5)
  - Regular buyers (20–45)

```python
print(df[['Age', 'Income', 'Purchases']].describe())
```

## 9. Perform Bivariate Analysis

```python
# Relationship: Gender vs Purchases
gender_purchase = df.groupby('Gender')['Purchases'].mean()
print(gender_purchase)
```

```python
gender_purchase.plot(kind='bar')
plt.title("Gender vs Average Purchases")
plt.xlabel("Gender")
plt.ylabel("Average Purchases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### What I Notices :
- Female avg ≈ 26.5  
- Male avg ≈ 27  
- Difference is negligible  
- Gender has **no impact** on purchases  
- Not useful for prediction  

```python
# Relationship: Income vs Churn
income_churn = df.groupby('Churn')['Income'].mean()
print(income_churn)
```

```python
income_churn.plot(kind='bar')
plt.title("Income vs Churn")
plt.xlabel("Churn (0=No, 1=Yes)")
plt.ylabel("Average Income")
plt.xticks(rotation=0) 
plt.show()
```

### What I Notices :

- Retained customers avg income ≈ ₹59,000  
- Churned customers avg income ≈ ₹50,500  
- Difference ≈ ₹8,500  
- Retained customers earn ~17% more  
- Lower-income customers churn more  
- Income is a **strong predictor of churn**  

###  10. Perform Multivariate Analysis

```python
import seaborn as sns
```

```python
# Correlation Heatmap
num_df = df[['Age', 'Income', 'Purchases']]
corr = num_df.corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")
plt.show()
```

### What I Notices 

- Age vs Income = 0.69 → strong positive  
- Income vs Purchases = 0.22 → weak positive  
- Age vs Purchases = 0.10 → very weak  
- Older customers earn more  
- Purchases weakly related to Age & Income  
- Age & Income show moderate multicollinearity  

```python
# Pair Plot
sns.pairplot(df, vars=['Age', 'Income', 'Purchases'], hue='Churn')
plt.show()
```

### What I Notices : 
#### Age
- Retained: spread across 35–40 and 50+  
- Churned: concentrated in 20–35  
- Younger customers churn more  

#### Income
- Retained: ₹60K–₹80K  
- Churned: ₹40K–₹55K  
- Clear separation → strong feature  

#### Purchases
- Retained customers purchase more  
- Churned customers limited to lower ranges  

---

### Scatter Plot Insights

### Age vs Income
- Strong positive trend  
- Young + low income → churned  
- Older + high income → retained  

### Income vs Purchases / Age vs Purchases
- Heavy overlap  
- No clear separation  
- Weak predictors individually
